```python
import os
import librosa
import soundfile as sf
from audiomentations import Compose, AddGaussianNoise, PitchShift, HighPassFilter, TimeStretch, Shift

# Define augmentation pipeline
augment_raw_audio = Compose([
    AddGaussianNoise(min_amplitude=0.0007, max_amplitude=0.005, p=0.5),
    PitchShift(min_semitones=-8, max_semitones=8, p=0.5),
    HighPassFilter(min_cutoff_freq=2000, max_cutoff_freq=4000, p=0.5),
    TimeStretch(min_rate=0.8, max_rate=1.25, p=0.5),
    Shift(p=0.5),
])

augment2 = Compose(
    [
        AddGaussianNoise(min_amplitude=0.0007, max_amplitude=0.010, p=0.5),
        PitchShift(min_semitones=-4, max_semitones=12, p=0.5),
        HighPassFilter(min_cutoff_freq=2000, max_cutoff_freq=3400, p=0.5)
    ]
)

augment3 = Compose(
    [
        AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.007, p=0.5),
        PitchShift(min_semitones=0, max_semitones=10, p=0.5),
        TimeStretch(min_rate=0.8, max_rate=1.25, p=0.5)
    ]
) 


# Define function to augment audio files
def augmentaudiofile1(working_dir, label):
    # Load audio files from the corresponding label folder
    folder = os.path.join(working_dir, label)
    
    for file in os.listdir(folder):
        file_path = os.path.join(folder, file)
        
        # Load audio file
        signal, sample_rate = librosa.load(file_path, mono=True, duration=4.5, sr=22050*2)
        
        # Augment audio
        augmented_signal = augment_raw_audio(signal, sample_rate)
        augmented_signal2 = augment2(signal, sample_rate)
        augmented_signal3 = augment3(signal, sample_rate)
        
        # Output directory for processed audio
        output_folder = os.path.join('/kaggle/working/outputaudio', label)  # Adjusted output folder path
        os.makedirs(output_folder, exist_ok=True)
        
        # Output file path
        output_file_path = os.path.join(output_folder, f'{file[:-4]}_processed.wav')
        output_file_path1 = os.path.join(output_folder, f'{file[:-4]}_processed1.wav')
        output_file_path2 = os.path.join(output_folder, f'{file[:-4]}_processed2.wav')
        output_file_path3 = os.path.join(output_folder, f'{file[:-4]}_processed3.wav')
        
        # Save the processed audio as a WAV file
        sf.write(output_file_path, augmented_signal, sample_rate, 'PCM_24')
        sf.write(output_file_path1, augmented_signal2, sample_rate, 'PCM_24')
        sf.write(output_file_path2, augmented_signal3, sample_rate, 'PCM_24')
        sf.write(output_file_path3, signal, sample_rate, 'PCM_24')

```

```python
# Define function to augment audio files
def augmentaudiofile2(working_dir, label):
    # Load audio files from the corresponding label folder
    folder = os.path.join(working_dir, f'{label}')
    
    for file in os.listdir(folder):
        file_path = os.path.join(folder, file)
        
        # Load audio file
        signal, sample_rate = librosa.load(file_path, mono=True, duration=4.5, sr=22050*2)
        
        # Augment audio
        augmented_signal = augment_raw_audio(signal, sample_rate)
        
        # Output directory for processed audio
        output_folder = os.path.join('/kaggle/working/outputaudio', label)  # Adjusted output folder path
        os.makedirs(output_folder, exist_ok=True)
        
        # Output file path
        output_file_path = os.path.join(output_folder, f'{file[:-4]}_processed.wav')
        output_file_path1 = os.path.join(output_folder, f'{file[:-4]}_processed1.wav')
        
        # Save the processed audio as a WAV file
        sf.write(output_file_path, augmented_signal, sample_rate, 'PCM_24')
        sf.write(output_file_path1, signal, sample_rate, 'PCM_24')

```

```python
labels1  = [    
    "Fart",
    "Shatter",
    "Hi-hat",
    "car_horn",
    "Splash_and_splatter",
    "Knock",
    "Laughter",
]

labels2  = [    
    "siren",
    "Snare_drum",
    "Guitar",
    "Gunshot_and_gunfire",
    "dog_barking",
    "drilling"
]
```

```python
# Call the function for each label
for label in labels1:
    augmentaudiofile1(working_dir, label)
```

```python
# Call the function for each label
for label in labels2:
    augmentaudiofile2(working_dir, label)
```
